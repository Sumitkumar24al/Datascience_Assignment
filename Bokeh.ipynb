Glyphs in Bokeh are the visual markers used to represent data points in a plot. They can be added to a Bokeh plot using the add_glyph() method of a figure

```python
x=np.array([1,2,3,4,5])
y=np.array([10,8,6,4,2])
p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")
p.circle(x,y,size=10,color='red',alpha=0.3)
show(p)
```

```python

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

p = figure(title="Customizing Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a circle glyph to the plot
p.circle(x, y, size=10, color="blue", alpha=0.8, legend_label="Data Points")

# Customize axes appearance
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12
p.xaxis.axis_label_text_font = "Arial"
p.yaxis.axis_label_text_font = "Arial"
p.xaxis.axis_line_color = "green"
p.yaxis.axis_line_color = "green"

# Customize title appearance
p.title.text_font_size = "16pt"
p.title.text_color = "navy"
p.title.align = "center"

# Customize legend appearance
p.legend.title = "Legend"
p.legend.label_text_font = "Helvetica"
p.legend.label_text_color = "black"
p.legend.label_text_font_size = "12pt"

show(p)
```

A Bokeh server is a Python process that allows for the creation of interactive web applications with dynamic, real-time updates. It enables the creation of interactive plots and data visualizations that can respond to user inputs, such as sliders, buttons, or dropdowns, and update the plot content in real time.

To use the Bokeh server to create interactive plots with real-time updates, you need to:

1. Create a Python script defining the plot and its interactive components, such as widgets.
2. Use the curdoc() function from bokeh.io to access the current document.
3. Define a callback function that updates the plot based on the changes in the interactive components.
4. Connect the callback function to the interactive components using the on_change or on_click methods.
5. Run the Bokeh server using the bokeh serve command, passing the Python script as an argument.



```python
from bokeh.plotting import figure, curdoc
from bokeh.models import Slider

p = figure(title="Interactive Plot", x_range=(0, 10), y_range=(0, 10))
p.line([1, 2, 3, 4, 5], [5, 4, 3, 2, 1], line_width=2)


slider = Slider(start=1, end=10, value=5, step=1, title="Slider")


def update_plot(attr, old, new):
    p.line([1, 2, 3, 4, 5], [new, new-1, new-2, new-3, new-4], line_width=2)

slider.on_change('value', update_plot)

curdoc().add_root(p)
curdoc().add_root(slider)
show(p)
```
