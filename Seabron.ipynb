Here are five commonly used plots that can be created using seaborn along with their uses:

1. Scatter Plot: To visualize the relationship between two continous variables and useful for identifying patterns, correlations, and outliers in the data.
2. Bar Plot: To display the relationship between a categorical variable and a continous variable and used to compare different categories and their corresponding values.
3. Histogram: To visualize the distribution of a single continuos variable and helps to understand the data's frequency distribution and detect potential outliers.
4. Box Plot: To display the distribution of data across multiple categories and provide insights into the spread, central tendency, and presence of the outliers in the data.
5. Heatmap: To visualize the correlation between multiple variables in a matrix form and useful for identifying patterns,clusters and relationship between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
```

```python
sns.lineplot(data=fmri,x=fmri.timepoint,y=fmri.signal,hue='event',style='region')
plt.title('FMRI signal overtime for different events and regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

```python
titanic=sns.load_dataset('titanic')
```

```python
plt.subplot(1,2,1)
sns.boxplot(data=titanic,x='pclass',y='age')
plt.title('Box Plot of Age vs Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1,2,2)
sns.boxplot(data=titanic,x='pclass',y='fare')
plt.title('Box Plot of Fare vs Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.show()
```

```python
diamonds=sns.load_dataset('diamonds')

sns.histplot(data=diamonds,x='price',hue='cut',kde=True,bins=25,alpha=0.5)
plt.show()
```

```python
diamonds=sns.load_dataset('diamonds')

sns.histplot(x=diamonds.price,data=diamonds,hue='cut',kde=True,bins=25,alpha=0.4)
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```
