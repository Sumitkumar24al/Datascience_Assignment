In probability theory, a Probability Density Function (PDF) is a mathematical function that describes the relative likelihood or probability of a continuous random variable taking on a specific value within a given range. It assigns a non-negative value to each point in the sample space, representing the probability per unit length for the variable to fall within that range. While the probability of the variable exactly equalling any single value is zero due to continuous nature, the PDF allows us to compare probabilities between different points in the sample space, helping us understand the likelihood of the variable being close to specific values. The PDF is crucial in analyzing and characterizing continuous probability distributions and is widely used in various statistical and data analysis applications.

There are several types of probability distributions, each associated with different types of random variables. The two main categories of probability distributions are:

1. Discrete Probability Distributions:

Bernoulli Distribution: Represents a single binary outcome (e.g., success or failure) with a fixed probability.
Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.
Poisson Distribution: Describes the number of events that occur in a fixed interval of time or space, given a constant average rate.

2. Continuous Probability Distributions:

Uniform Distribution: Assumes all values within an interval are equally likely.
Normal/Gaussian Distribution: Characterized by a symmetric bell-shaped curve and is widely used due to the Central Limit Theorem.
Exponential Distribution: Models the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x,mean,std):
    pdf=(1/(std*np.sqrt(2*np.pi)))*np.exp(-(x-mean)**2/(2*std**2))
    return pdf

def plot_normal_pdf(mean,std):
    x=np.linspace(mean-3*std,mean+3*std,1000)
    pdf_values=normal_pdf(x,mean,std)
    
    plt.plot(x,pdf_values,label=f"mean={mean}, std_dev={std}")
    plt.xlabel('x')
    plt.ylabel('Probability density')
    plt.title('Normal Distribution pdf')
    plt.legend()
    plt.grid(True)
    plt.show()
    
```

```python
mean=0
std=1
plot_normal_pdf(mean,std)
```

The properties of the Binomial distribution are as follows:

1. Fixed Number of Trials: 
The Binomial distribution requires a fixed number of trials, denoted as "n", where each trial is independent of the others.

2. Constant Probability: 
The probability of success in each trial remains constant and is denoted as "p". The probability of failure in each trial is 1 - p.

3. Discrete Distribution: 
The Binomial distribution is a discrete probability distribution, meaning it deals with whole number outcomes (0, 1, 2, ...).

4. Probability Mass Function: 
The probability mass function (PMF) for the Binomial distribution is given by:
P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

where X is the number of successes, k is the specific number of successes, C(n, k) is the binomial coefficient (number of combinations), and p^k and (1 - p)^(n - k) are the probabilities of getting exactly k successes and (n - k) failures, respectively.

5. Mean and Variance: 
The mean (expected value) of the Binomial distribution is n * p, and the variance is n * p * (1 - p).

Two examples where the Binomial distribution can be applied are:

1. Coin Tossing: 
Consider a fair coin being tossed 10 times. Each toss has two possible outcomes (heads or tails) with a fixed probability of 0.5 (fair coin). The number of heads obtained in these 10 tosses can be modeled using the Binomial distribution.

2. Product Quality Control: 
A manufacturing company produces electronic components, and each component can be either defective or non-defective. The probability of producing a defective component is known (e.g., 0.05). The company conducts quality control by inspecting a fixed number of components, say 100, and records the number of defective components found. The number of defective components in the sample can be modeled using the Binomial distribution.

```python
from scipy.stats import poisson

def poisson_cdf(point,mean):
    cdf_value=poisson.cdf(point,mean)
    return cdf_value
```

```python
mean=3.5
x=np.arange(0,21)
cdf_value=[poisson_cdf(point,mean) for point in x]
print(f'CDF at point {point} for Poisson distribution with mean {mean} : {cdf_value}')
plt.plot(x,cdf_value,marker='o', linestyle='-', color='b')
plt.xlabel('Number of Successes')
plt.ylabel('Cumulative Probability')
plt.title(f'CDF of Poisson Distribution (mean={mean})')
plt.grid(True)
plt.show()
```

Ques 7

The Binomial distribution and Poisson distribution are both probability distributions used to model the number of successes or occurrences in a series of independent trials. The key distinction lies in the context and assumptions of their applications.

The Binomial distribution is employed when there is a fixed number of trials (n) with two possible outcomes (success or failure) in each trial. It assumes that the trials are independent, and the probability of success (p) remains constant across all trials. This distribution is commonly used for scenarios like coin tosses, pass-fail exams, or quality control processes where the focus is on the number of successful outcomes in a fixed set of trials.

On the other hand, the Poisson distribution is employed when the number of trials is not fixed, and the interest lies in modeling rare events or occurrences in a fixed interval of time or space. It represents the number of events that happen within the interval, given a constant average rate of occurrence (λ). The Poisson distribution assumes that events occur randomly and independently over time or space, making it suitable for scenarios like modeling the number of accidents in a day, the number of customers arriving at a store in an hour, or the number of defects in a product.

```python
mean=5
sample_size=1000

random_sample=np.random.poisson(mean,size=sample_size)
sample_mean=np.mean(random_sample)
sample_variance=np.var(random_sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)

plt.hist(random_sample, bins=20, alpha=0.6)
plt.xlabel('Number of Occurrences')
plt.ylabel('Frequency')
plt.title(f'Histogram of Random Sample from Poisson Distribution (mean={mean})')
plt.grid(True)

plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(sample_variance, color='blue', linestyle='dashed', linewidth=2, label='Sample Variance')
plt.legend()
plt.show()
```

Ques 9

the relationship between mean and variance differs in the Binomial and Poisson distributions. In the Binomial distribution, the mean (μ) is the product of the number of trials (n) and the probability of success in each trial (p), while the variance (σ^2) is determined by both the number of trials and the probability of success and failure. Specifically, the variance is given by σ^2 = n * p * (1 - p). As the probability of success varies or the number of trials changes, the variance also changes accordingly.
       On the other hand, in the Poisson distribution, both the mean and variance (λ) are equal. The variance is solely determined by the average rate of occurrence, and it remains constant regardless of any other factors. In both distributions, an increase in the mean typically corresponds to an increase in the variance, indicating a broader spread of data around the central value. Understanding the relationship between mean and variance is essential for interpreting and analyzing data represented by these probability distributions.
       
       In both distributions, the variance is directly related to the mean. In the Binomial distribution, the variance depends on both the number of trials (n) and the probability of success (p), while in the Poisson distribution, the variance is solely determined by the average rate of occurrence (λ). As a general trend, in both distributions, as the mean increases, the variance also tends to increase, indicating a greater spread of the data around the mean.

In a normal distribution, the least frequent data points are located in the tails of the distribution, which are the regions farthest from the mean. The normal distribution is symmetric, and the data is centered around the mean. As the data moves away from the mean in both positive and negative directions, the frequency of occurrence decreases. Within one standard deviation from the mean, approximately 68% of the data lies, and within two and three standard deviations, approximately 95% and 99.7% of the data lie, respectively. Data points beyond three standard deviations from the mean are considered rare and account for only a small percentage of the total data. Hence, the least frequent data points in a normal distribution are the ones that are several standard deviations away from the mean.
