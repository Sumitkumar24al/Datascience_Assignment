Scatter Plot is a type of data visualization that is used to visualize the relationship between two continous variables by displaying individual data points as dots on 2-D plane. Each dot on the plot represents a single data observation, and its position on the graph corresponds to the values of the two variables for that observation. They are paticularly useful for identifying patterns, trends and correlations between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))

plt.scatter(x,y)
plt.title('Scatter Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

The subplot() function in MAtplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots(subplots) in a single figure, arranging them in rows and columns.
It takes 3 arguments: nrows(the number of rows in the subplot grid), ncolumns(the no. of columns in subplot grid) and index(index of current subplot that we want to create)

```python
import numpy as np
# For line 1:
x=np.array([0,1,2,3,4,5])
y=np.array([0,100,200,300,400,500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Subplot 1')
plt.xlabel('x')
plt.ylabel('y')

# For line 2:
x=np.array([0,1,2,3,4,5])
y=np.array([50,20,40,20,60,70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('Subplot 2')
plt.xlabel('x')
plt.ylabel('y')

# for line 3:
x=np.array([0,1,2,3,4,5])
y=np.array([10,20,30,40,50,60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('Subplot 3')
plt.xlabel('x')
plt.ylabel('y')
           
# For line 4
x=np.array([0,1,2,3,4,5])
y=np.array([200,350,250,550,450,150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('Subplot 4')
plt.xlabel('x')
plt.ylabel('y')

plt.subplots_adjust(left=0.1,right=0.9,bottom=0,top=0.9,wspace=0.4,hspace=0.4)
plt.show()
```

A bar plot is a data visualization technoque used to represent categorical data with rectangular bars. Each bar typically represents a specific category, and the length or height of the bar corresponds to the value or frequency of that category. Bar plots are widely used for comapring different categories or groups, showing the distribution of data, and visualizing trends.

Bar plots are commonly used for :
1. Comparison between different categories or groups.
2. For displaying categorical data, where each category is distinct and does not have a natural order.
3. Suitable for displaying discrete data points or data grouped in categories.
4. Used to display the frequency or occurrence of different categories in a dataset.


```python
company=np.array(['Apple','Microsoft','Google','AMD'])
profit=np.array([3000,8000,1000,10000])
plt.barh(company,profit)
plt.title('Company profits')
plt.xlabel('Company')
plt.ylabel('Profits')
plt.show()
```

Box plots is a data visualization technique used to display the distribution and key statistical properties of a dataset. It summarizes the data's central tensency(mean,median,mode), dispersion(variance,std.deviation) and presence of outliers. Box plots are particularly useful for comparing the distribution of the multiple datasets or identifying the presence of outliers in a single dataset. 

Box plots are used for following reasons:
1. It provides a visual representation of the data's spread and distribution.
2. It is used for comparing multiple datasets.
3. It can identify potential outliers in the datasets.
4. It provide insights into the symmetry and asymmetry of the data distribution.
6. It allows the comparison of different datasets in terms of central tendency, spread, and presence of outliers.

```python
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
data=[box1,box2]
plt.boxplot(data)
plt.xticks([1,2],['box1','box2'])
plt.title('Box Plot')
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.show()
```
